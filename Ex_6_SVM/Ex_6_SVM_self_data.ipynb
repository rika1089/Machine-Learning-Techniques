## To perform SVM on given Dataset 

____
## step 0 : Import  the necessary libraries 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report,accuracy_score
from sklearn.svm import SVC
from sklearn.feature_extraction.text import TfidfVectorizer
```

____
## sample dataset (emails &amp ; labels : 1 for spam nad : 0 for legit)



```python
emails = ["Win a lottery now", "Meeting scheduled for tomorrow", "Get discount on medicines", 
          "Your bank account is updated", "Urgent: Update your password"]
labels = [1, 0, 1, 0, 1]  # Spam = 1, Not Spam = 0
 
```

### Convert text data into mathematical form using Tfidf Vectorizer

```python
vectoriser = TfidfVectorizer()
X = vectoriser.fit_transform(emails)
```

# Splitting the dataset into training and testing


```python
X_train,X_test,y_train,y_test = train_test_split(X,labels,test_size=0.2, random_state=42)
```


### Train the SVM model


```python
# Train the SVM model
svm_model = SVC(kernel='linear', C=1.0)
svm_model.fit(X_train, y_train)
 
# Make predictions
y_pred = svm_model.predict(X_test)
 
# Evaluate performance
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
 
print("Model Accuracy:", accuracy)
print("Classification Report:\n", report)
```
